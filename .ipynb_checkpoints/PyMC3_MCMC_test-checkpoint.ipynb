```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pymc3 as pm
from pymc3 import glm, plot_posterior_predictive_glm
import arviz as az

# AstroML package
from astroML.datasets import simulation_kelly
from astroML.linear_model import LinearRegressionwithErrors
from astroML.plotting import plot_regressions, plot_regression_from_trace

import warnings
warnings.filterwarnings('ignore')
```

### AstroML Implementation

```python
ksi, eta, xi, yi, xi_error, yi_error, alpha_in, beta_in = simulation_kelly(size=100, 
                                                                           scalex=0.2, 
                                                                           scaley=0.2,
                                                                           alpha=2, 
                                                                           beta=1, 
                                                                           epsilon=(0, 0.75))
```

```python
ksi_0 = np.arange(np.min(ksi) - 0.5, np.max(ksi) + 0.5)
eta_0 = alpha_in + ksi_0 * beta_in

figure = plt.figure(figsize=(10, 8))
figure.subplots_adjust(left=0.1, right=0.95,
                       bottom=0.1, top=0.95,
                       hspace=0.1, wspace=0.15)
ax = figure.add_subplot(221)
ax.scatter(ksi, eta)
ax.set_xlabel(r'$\xi$')
ax.set_ylabel(r'$\eta$')

ax.plot(ksi_0, eta_0, color='orange')

for scalex, scaley, axn in [(0.1, 0.1, 222), (0.3, 0.5, 224), (0.2, 0.2, 223)]:
    _, _, xi, yi, xi_error, yi_error, _, _ = simulation_kelly(size=100, scalex=scalex, scaley=scaley,
                                                              alpha=alpha_in, beta=beta_in, ksi=ksi, eta=eta)    
    ax = figure.add_subplot(axn)

    ax.scatter(xi[0], yi, alpha=0.5)
    ax.errorbar(xi[0], yi, xerr=xi_error[0], yerr=yi_error, alpha=0.3, ls='')
    ax.set_xlabel(f'x, error scaler: {scalex}')
    ax.set_ylabel(f'y, error scaler: {scaley}')

    x0 = np.arange(np.min(xi) - 0.5, np.max(xi) + 0.5)
    y0 = alpha_in + x0 * beta_in
    ax.plot(x0, y0, color='orange')
```

```python
linreg_xy_err = LinearRegressionwithErrors()
linreg_xy_err.fit(xi, yi, yi_error, xi_error)
```

```python
plot_regressions(ksi, 
                 eta, 
                 xi[0], 
                 yi, 
                 xi_error[0], 
                 yi_error, 
                 add_regression_lines=True, 
                 alpha_in=alpha_in, 
                 beta_in=beta_in)

plot_regression_from_trace(linreg_xy_err, 
                           (xi, yi, xi_error, yi_error), 
                           ax=plt.gca(), 
                           chains=50)
```

Likelihood Function (1)

$$\large{\hat{x}_i = \hat{\xi}_i + \hat{\epsilon}_{x,i}}$$

Likelihood Function (2)

$$\large{\vec{y}_i = \vec{\eta}_i + \vec{\epsilon}_{y,i} = \alpha + \vec{\beta} \cdot \hat{\xi}_i + \vec{\epsilon}_{y,i}}$$

Given, $$\large{\vec{\eta}_i = \alpha + \vec{\beta} \cdot \hat{\xi}_i}$$

```python
# model object
model = pm.Model()
# instance of model object
with model:
    
    # ----PRIORS----
    alpha = pm.Normal('alpha', mu=0, sd=10)
    beta = pm.Normal('beta', mu=0, sd=10)
    
    # ----2 LIKELIHOOD FUNCTIONS----
    # xi = ksi  with a difference of xi_error
    xi_pm = pm.Normal('xi_pm',
                      mu=ksi, 
                      sd=xi_error, 
                      shape=xi.shape, 
                      observed=xi)
    
    # yi = eta with a difference of yi_error
    # eta = alpha + beta*ksi
    yi_pm = pm.Normal('yi_pm', 
                      mu=alpha + beta*xi_pm, 
                      sd=yi_error, 
                      shape=eta.shape, 
                      observed=yi)
    
    # ----SAMPLING----
    trace = pm.sample(draws=5000,
                      tune=50000,
                      chains=3,
                      return_inferencedata=False)
```

```python
trace
```

```python
az.summary(trace)
```

```python
plot_regressions(ksi, 
                 eta, 
                 xi[0], 
                 yi, 
                 xi_error[0], 
                 yi_error, 
                 add_regression_lines=True, 
                 alpha_in=alpha_in, 
                 beta_in=beta_in)
print()
print('...AstroML regression method results...')
plot_regression_from_trace(linreg_xy_err, 
                           (xi, yi, xi_error, yi_error), 
                           ax=plt.gca(), 
                           chains=50)

alpha_mean = trace['alpha'].mean()
beta_mean = trace['beta'].mean()
plt.plot(xi[0], (alpha_mean + beta_mean*xi)[0], color='gold', label='MCMC predictions')
plt.legend()
print()
print('...MCMC regression method results...')
print('beta val:', beta_mean, 'alpha val:', alpha_mean)
```
